```python
import requests
from bs4 import BeautifulSoup
```

```python
douban_url = 'https://book.douban.com/top250?start=0'
bilibili_url = 'https://www.bilibili.com/v/popular/rank/all'
```

```python
my_headers = my_headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/101.0.4951.54 Safari/537.36"}
result = requests.get(douban_url, headers=my_headers)   # 发起网络请求
print("result: ", result)  # 返回200代表成功
print("status_code: ", result.status_code)
```

```python
print(result.text)  # 打印返回结果
```

```python
html_text = result.text
type(html_text)
```

```python
soup = BeautifulSoup(html_text, 'html.parser')  # 将html文本传给beautifulsoup并告诉它这是html格式
```

```python
print(soup.title.text)  # 打印标题
```

```python
items = soup.find_all('table')
print(len(items))
```

```python
items2 = soup.find_all('table',{'width':"100%"})  # 通过字典细化查找
print(len(items2))
```

```python
from pyecharts.charts import Map
from pyecharts import options as opts 
```

```python
nations = ["China", "Russia", "Germany", "South Africa", "Canada"]
values = [95.1, 5, 12, 20, 10]
data = [   tuple(element)   for element in zip(nations, values) ]
```

```python
map = Map()
map.add("世界地图",data, "world")
map.set_series_opts(label_opts=opts.LabelOpts(is_show=False))
map.set_global_opts(
    title_opts=opts.TitleOpts(title="各国旅游人数", subtitle="数据来源：纯属杜撰", pos_right="center", pos_top="5%"),
    visualmap_opts=opts.VisualMapOpts(max_=100, min_=1)
    )
map.render_notebook() 
```
