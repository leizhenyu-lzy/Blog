```python
def angle2color(angle):
    """
    将角度 [-π, π] 映射到 RGB 颜色

    HSV 色彩空间 (OpenCV 范围)
    1. Hue          色  相   : 0 - 179      (颜色的种类，例如红色、绿色、蓝色等)
    2. Saturation   饱和度   : 0 - 255      (颜色的纯度或浓度 : 饱和度越高，颜色越鲜艳；饱和度越低，颜色越灰暗，趋于无色)
    3. Value        亮  度   : 0 - 255      (颜色的明亮程度 : 值越高，颜色越亮；值越低，颜色越暗)
    """
    # 将角度从 [-π, π] 映射到 [0, 1]
    normalized_angle = (angle + np.pi) / (2 * np.pi)

    # 将 [0, 1] 映射到 [0, 179]，表示 HSV 中的 Hue 值（OpenCV 中 Hue 的范围是 0 到 179）
    hue = np.round(normalized_angle * 179).astype(np.uint8)

    # 将 HSV 转换为 RGB
    hsvColor = np.array([hue, 255, 255], dtype=np.uint8).reshape((1,1,3))

    print("hsvColor : ", hsvColor.flatten())

    rgbColor = cv2.cvtColor(hsvColor, cv2.COLOR_HSV2BGR).reshape(-1)  # 转换为 RGB 颜色

    print("rgbColor : ", rgbColor)

    return rgbColor

```

```python
rgb = angle2color(-np.pi)
rgb = angle2color(0)
rgb = angle2color(+np.pi)
```
